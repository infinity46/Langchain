### Ollama

 Ollama supports embeddings models, making it possible to build retrieval augumented generation (RAG) application that combine text prompts with existing documents or other data

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma:2b")     #by default it uses lamma2
)
```

```python
embeddings
```

```python
r1=embeddings.embed_documents(
    [
        "Alpha is the first letter of the Greek alphabet.",
        "Beta is the second letter of the Greek alphabet."
    ]
)
```

```python
len(r1)
```

```python
embeddings.embed_query("what is the second letter of the Greek alphabet?")
```
