```python
## Open source models -- LLama3, Gemma2, mistral -- groq
import os 
from dotenv import load_dotenv
load_dotenv()

groq_api_key=os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq

model=ChatGroq(model="llama-3.1-8b-instant", groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage(content="Translate the following from english to German"),
    HumanMessage(content="Hello how are you?")
]

result=model.invoke(messages)
```

```python
result
```

```python
from langchain_core.output_parsers import StrOutputParser

parser=StrOutputParser()
parser.invoke(result)
```

```python
## Using LCEL - chain the components 

chain=model | parser
chain.invoke(messages)
```

```python
### Prompt Template

from langchain_core.prompts import ChatPromptTemplate

generic_template="Translate the following into {language}: "

prompt=ChatPromptTemplate.from_messages(
    [("system", generic_template),("user", "{text}")]
)

```

```python
result=prompt.invoke({"language":"German", "text":"Hello how are you?"})
```

```python
result.to_messages()
```

```python
##chaining togerther components with LCEL

chain = prompt | model | parser
chain.invoke({"language":"German", "text":"Good morning, how are you?"})
```
